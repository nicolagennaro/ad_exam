```python
%matplotlib inline
import numpy as np
from pylab import *
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi']= 200 
```

```python
kk = [100000,200000,300000,400000, 500000, 600000, 700000, 800000, 900000, 1000000]#, 12000000]
medie_before = []
errori_before = []
for j,kj in enumerate(kk):
    m = []
    with open('/home/nicole/Data Science/ad_exam/c++/test/output_before_'+str(kj)) as file:
            for line in file:
                m.append(float(line))
    medie_before.append(np.mean(m))
    errori_before.append(np.std(m))
    
print(medie_before)

medie_after = []
errori_after = []
for j,kj in enumerate(kk):
    m = []
    with open('/home/nicole/Data Science/ad_exam/c++/test/output_after_'+str(kj)) as file:
            for line in file:
                m.append(float(line))
    medie_after.append(np.mean(m))
    errori_after.append(np.std(m))
    
print(medie_after)
```

```python
fig, ax = plt.subplots(figsize=(7, 4))
plt.legend(bbox_to_anchor=(0.5, 1.02, 1., .102), loc=3,alpha=0.1,
           ncol=2, mode="expand", borderaxespad=0.,fontsize=4,fancybox=True,shadow=False)
x = np.array(kk)
#ax.set_yscale("log")
#ax.set_xscale("log")
ax.errorbar(x, medie_before, yerr=errori_before ,linewidth=0.8, label='before using balance function')
ax.errorbar(x, medie_after, yerr=errori_after ,linewidth=0.8, label='after using balance function')
ax.set_xticks(x)
#ax.set_xticklabels(['1','2','4','8','16','32','64','128','256','512','1024'])
ax.set_title('Times for find function',fontsize=10)
plt.xlabel('Tree size (N)')
plt.ylabel('Time(s)')
ax.grid(alpha=0.2,linestyle='--')

plt.rc('legend',**{'fontsize':8})
plt.legend()

plt.show()
```
